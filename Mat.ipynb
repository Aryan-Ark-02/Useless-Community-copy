```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
pd.set_option('future.no_silent_downcasting', True)
```

```python
data = pd.read_csv('sharma-kohli.csv')
df = pd.read_csv('batter.csv')
```

## Line Chart

```python
plt.plot(data['index'],data['RG Sharma'],label='Virat',color='#39CCCC',linestyle='solid',linewidth=1,marker='d')
plt.plot(data['index'],data['V Kohli'],label='Rohit',color='#FF7F50',linestyle='dashdot',marker='+')
plt.title('Rohit Sharma And Virat Kohli Comparision')
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.grid()
plt.legend()
;
```

## Scatter Plot

```python
df
plt.scatter(df['strike_rate'],df['avg'],marker='+',color='#001f3f')
plt.scatter(df['avg'],df['strike_rate'],marker='+',color='#FFDD57')
```

## Bar Chart

```python
plt.bar(data['index'],data['RG Sharma'],color=['#DC143C'])
plt.title('Rohit Sharma')
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.grid(True,axis='y')
```

```python
plt.bar(data['index'],data['V Kohli'],color=['#673AB7'])
plt.title('Virat Kohli')
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.grid(True,axis='y')
```

## **Stacking Bar**

## This is Very Bad 👇

```python
plt.bar(data['index'],data['RG Sharma'],color='#B87333')
plt.bar(data['index'],data['V Kohli'],color='#5D8AA8')
plt.title('Rohit Sharma And Virat Kohli (WRONG WAY)')
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.grid(True,axis='y')
```

## This is the correct way 👇

```python
plt.bar(data['index'],data['RG Sharma'],label='RG Sharma',color='lightblue')
plt.bar(data['index'],data['V Kohli'],label='V Kohli',color='pink',bottom=data['RG Sharma'])
plt.title('Rohit Sharma And Virat Kohli Comparision')
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid(True,axis='y')
plt.xticks(rotation=390)
```

```python
f = pd.read_csv('iris.csv')
f.replace({'Species' : {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}},inplace=True)
plt.figure(figsize=(15,5))
plt.scatter(f['SepalLengthCm'],f['PetalLengthCm'],c=f['Species'],s=f['PetalWidthCm']*120,marker='o',alpha=0.9)
plt.colorbar()
plt.scatter(f['SepalWidthCm'],f['PetalWidthCm'],c=f['Species'],s=f['PetalWidthCm']*100,marker='^',cmap='cool',alpha=0.4)
plt.colorbar()
plt.title('Iris')
plt.xlabel('Length in Cm')
plt.ylabel('Length in Cm')
plt.text(6.5,1,'△ Represents Sepal')
plt.text(6.5,1.5,'○ Represents Petal')
plt.axvline(4.7)
plt.axhline(3)
plt.savefig('mew.png')
;
```
