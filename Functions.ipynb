#### Reciprocal

```python
x=np.linspace(-10,10,100)
y = list(map(lambda x: (1/x),x))
plt.plot(x,y,color='gold')
plt.title('Plot of f(x) = 1/x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
```

```python
x=np.linspace(0,np.pi*4,100)
y=np.sin(x)
z=np.cos(x)

plt.plot(x,y,color='red',label='Sine')
plt.plot(x,z,color='blue',label='Cos')
plt.title('Sine/Cosine Graph')
plt.xlabel('Radian')
plt.ylabel('Sin(x)/Cos(x) value')
plt.axhline(color='brown')
plt.grid()
plt.legend();
```
